## Barplot :

### Here we are checking that what are the most popular neighborhoods for Airbnb listings in NYC.

    We are using here column "Neighbourhood_group" and after grouping the data based on this column we are dividing the 
    groups on basis of "Room_type" and and then checking their "value_counts". 

```python
print(data1.groupby("Neighbourhood_Group")["Room_Type"].value_counts())

plt.figure(figsize=(8,4))
sns.countplot(data=data1,x="Neighbourhood_Group",hue="Room_Type")
plt.xticks(fontsize=9)
plt.title("Most Popular neighbourhood",fontsize=13.5,color="green")
plt.xlabel(xlabel=None,loc=None)
plt.legend(fontsize="small",loc='upper right')
plt.show()
```

    From here we can check that what is the most Popular neighbourhood for Airbnb listings in NYC.Some conclusions from here
    1. It is quite clear from here that for 'Private rooms' both Manhattan and Brooklyn are equally favourite.
    2. For entire Home or apartment 'Manhattan' is more favourite thatn brooklyn.
    3. In Manhattan shared rooms are also available.

## Categorical Plot :

### Here we are checking for Price of rooms in different neighbourhood.

    We are using here "Neighbourhood_Group" and "Price($)" columns to draw a Categorical Plot. Here we further devide those
    categories on the basis of "Room_Type"

```python
"""From here we can draw conclusion that all the prices are fare, and there is one price which is extreme, but it doesn't 
seems like outlier."""

sns.catplot(data=data1,x="Neighbourhood_Group",y="Price($)",hue="Room_Type",height=5.5,aspect=1.3)
plt.title("Price of rooms based on neighbourhood",fontsize=13.5,color="green")
plt.xlabel(xlabel=None,loc=None)
plt.show()
```

## Pie Chart : 

### Here we are checking for Type of room with most reviews. 

    We are using here calculating "Number_Of_Reviews" with respect to "Room_Type" to get the idea about most popular room type.

```python
print(data1.groupby("Room_Type")["Number_Of_Reviews"].sum())

plt.pie(x=data1.groupby("Room_Type")["Number_Of_Reviews"].sum().tolist(),
        labels=data1.groupby("Room_Type")["Number_Of_Reviews"].sum().index,
       autopct='%1.1f%%',colors=["C8","C9","r"])
```

    From here we can draw the conclusion that most of the reviews are for private room only. That means more people are
    using Private rooms.

## Violin plot :

### Here we are trying to look for Rooms availability based on neighbourhood.

    We are checking here that in which city rooms are available for most of the year. Here we are using Violin plot
    for that purpose. We further divide these violin plots into subcategories of different room types

```python
print(data1.groupby(["Neighbourhood_Group","Room_Type"])["Availability_365"].count())

plt.figure(figsize=(12,5))
sns.violinplot(data=data1,x="Neighbourhood_Group",y="Availability_365",hue="Room_Type")
plt.title("Reviews for Rooms based on neighbourhood",fontsize=13.5,color="green")
plt.xlabel(xlabel=None,loc=None)
plt.show()
```

    From here we can draw the conclustion that, Manhattan each type of rooms are available for most of the year

## Histogram :

### Here we are looking for How many employee are getting hired each year.

    For that we are using "DateofHire" columnn and counting the number of empoyees in that year.

```python
# Number of Employees hired each year

plt.figure(figsize=(7,5))
ax=sns.histplot(x=data2["DateofHire"])
plt.title("Number of employee hired each year",fontsize=13,color="green")
plt.ylim(0,80)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x()+0.4, p.get_height()+5))
plt.show()
```

    From Here we can draw the conclusion that in 2012 max employee got hired.

## Scatter Plot :

### Here we are looking for Saleries of both male and female.

    For that we are using "Salary" column and categorizing them according to their Gender.

```python
# Looking for salary recieved by most Male and Females

plt.figure(figsize=(8,6))
plt.figure(figsize=(12,4))
sns.scatterplot(data=data2,x=data2.index,y="Salary",hue="Sex")
plt.title("Salaries for Males and Females",fontsize=13,color="green")
plt.legend(bbox_to_anchor=(1.1,1))
plt.show()
```

## Box plot :

### Here we are looking for Salaries of employee working in different departments.

    For that we are using "Salary" of employees and categorizing them according to their "Department"

```python
# Department wise salary

print(data2.groupby("Department")["Salary"].agg([np.min, np.mean, np.max]))

plt.figure(figsize=(12,5))
sns.boxplot(data=data2,x="Department",y="Salary")
plt.title("Department wise salary",fontsize=13,color="green")
plt.show()
```

    From here we can draw the conclusion that employee from "IT" department are having good salary
